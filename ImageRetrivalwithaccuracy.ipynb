```python
import numpy as np 
from PIL import Image
import os
```

### load image, resize image, create label based on filename

```python
def load_image_from_folder(folder):
    images = []
    filenames = []

    category, idx = "none", 0
    for filename in sorted(os.listdir(folder)):
        if filename.endswith(".png"):
            image = Image.open(os.path.join(folder, filename))
            image = image.resize((32, 32), Image.ANTIALIAS)
            if image is not None:
                images.append(image)
                filenames.append(filename)    
    return images, filenames
```

```python
folder = "/content/coil-20-proc"
images, filenames = load_image_from_folder(folder)
print(len(images), len(filenames))
```

```python
from IPython.display import display
for i in range(5):
    display(images[i])
```

### Convert each image to a vector and store whole dataset in a 2d numpy array

```python
tmp = []
for image in images:
    im = np.asarray(image)
    im = im.reshape((1024,))
    tmp.append(im)
imgs_array = np.array(tmp)
y = np.array(filenames)
```

### Splite dataset into Training (gallery) and Testing (Query/Probe)

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(imgs_array, y, test_size=0.2, random_state=0)
```

### Testing: for each probe image
- Calculate the distance to all gallery images
- Rank all the distance
- Display the most similar images

```python
from sklearn.metrics.pairwise import euclidean_distances
accuracy_overall = []
for idx, (query,query_filename) in enumerate(list(zip(X_test,y_test))):

  query = query.reshape((1, -1))
  D = euclidean_distances(X_train, query).squeeze()
  index = np.argsort(D)
  index = index[:10]

  tmp = [query.reshape((32,32))]

  for ind in index:
      tmp.append(X_train[ind].reshape((32,32)))
  output = np.array(tmp)*255
  output = output.transpose(1, 0, 2)
  output = output.reshape((32, -1))
  im_query = Image.fromarray(output)
  display(im_query)

    
  query_obj = query_filename.split("__")[0]
    
  tmp_filenames = []

  for new in index:
      imagee = y_train[new]
      tmp_filenames.append(imagee)
  
  filenames_only = [y.split("__")[0] for y in tmp_filenames]
  


  precision = 0

  for z in filenames_only:
    if z == query_obj:
     precision+=1
    else:
     pass
  accuracy = precision*10
  print("Accuracy for this first test image: ",str(accuracy) + "%")
  accuracy_overall.append(accuracy)


```

```python
#Calculating overall accuracy  - Average
acc = np.array(accuracy_overall)
print("Accuracy Overall: ",acc.mean())
```
